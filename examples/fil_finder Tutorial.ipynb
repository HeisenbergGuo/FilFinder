fil_finder Tutorial
===================

Load in the algorithm and the usual suspects.

```python
from astropy.io.fits import getdata
from fil_finder import fil_finder_2D
import matplotlib.pyplot as p
%matplotlib inline
```

First, load in the FITS file containing the image.

```python
img, hdr = getdata("filaments_updatedhdr.fits", header=True)
```

Next, initialize fil_finder_2D. 
The algorithm requires a few inputs to begin (other than the image and header):
        *  beamsize in arcseconds (= 15.1)
        *  skeleton threshold - minimum pixels a skeleton must contain to be considered (= 30)
        *  branch threshold - minimum length for a branch. If the intensity along it is significant, it will still be kept. (= 5)
        *  pad size - number of pixels to pad around each filament. Must be at least 1. (= 50)
        *  flatten threshold - percentile of the image to set the "elbow" of the arctan transform to. (= 95)
        *  distance - distance to the region in parsecs. Used to set the size of the adaptive thresholding patch. (= 260)
        *  global threshold - sets the percentile of data to ignore. Useful for not considering noise regions. (= 20)

```python
fils = fil_finder_2D(img, hdr, 15.1, 30, 5, 50, 95 ,distance=260, glob_thresh=20)
```

The algorithm has several steps, which will be outlined below. Using the run() function will perform all the steps in one call.

We begin by creating the mask of the image. 

```python
fils.create_mask(verbose=True)
```

A preview of the mask is shown above. If this is run in ipython's pylab mode, one can change the vmax on the image to inspect the mask over a wide range of intensities.

Pre-made masks can also be supplied to the algorithm without completing this step.

The Medial Axis Transform returns the skeleton objects.

```python
fils.medskel(verbose=True)
```

Now begins the analysis of the filaments! This begins with finding the length. Skeletons are also pruned during this process to remove short branches which aren't essential.

```python
fils.analyze_skeletons(verbose=True)
```
